```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
more_data = pd.read_csv("Dig-mnist.csv")
```

```python
data
```

```python
data.describe()
```

```python
digits = data.drop(['label'],1).values
digits2 = more_data.drop(['label'],1).values
```

```python
digits
digits2
```

```python
digits = digits/ 255
digits2 = digits2/255
label = data['label'].values
label2 = more_data['label'].values
```

```python
fig, axis = plt.subplots(5, 5, figsize=(22, 20))
for i, ax in enumerate(axis.flat):
    ax.imshow(digits[i].reshape(28, 28))
    ax.set(title = "Real digit is {}".format(label[i]))
```

```python
fig, axis = plt.subplots(5, 5, figsize=(22, 20))
for i, ax in enumerate(axis.flat):
    ax.imshow(digits2[i].reshape(28, 28))
    ax.set(title = "Real digit is {}".format(label2[i]))
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
digits.shape
```

```python
digits2.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(digits,label,test_size = 0.02)
x_train2,x_test2,y_train2,y_test2 = train_test_split(digits2,label2,test_size = 0.02)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
x_train
```

```python
x_train2
```

```python
x_train = np.concatenate((x_train, x_train2))
```

```python
y_train = np.concatenate((y_train, y_train2))
```

```python
model = RandomForestClassifier(n_estimators=1000, max_depth=10, min_samples_split=10)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print("Model accuracy is: {0:.3f}%".format(accuracy_score(y_test, y_pred) * 100))
```

```python
fig, axis = plt.subplots(5, 5, figsize=(18, 20))

for i, ax in enumerate(axis.flat):
    ax.imshow(x_test[i].reshape(28, 28), cmap='binary')
    ax.set(title = "Predicted digit {0}\nTrue digit {1}".format(y_pred[i], y_test[i]))
```

```python
test_data1 = test_data.drop(['id'],1)
test_x = test_data1.values
```

```python
my_preds = model.predict(test_x)
```
